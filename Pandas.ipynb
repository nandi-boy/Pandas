Pandas is an open source data analysis library written in python

Pandas has two types of data structures:

a) Dataframe: It's a tabular spreadsheet like structure representing rows each of which contains one or multiple columns.

b) Series: It's a one dimentional array with indexes,it stores a single column or row of data in a Dataframe.

A one-dimentional array(labeled) capable of holding any type of data-Series.

A two-dimentional data(labeled) structure with columns of potentially different types of data-DataFrame.
